```python
import folium
import numpy as np
import pandas as pd
import os

base_dir = '/kaggle/input/'

def plot_points_on_map(dataframe,
                       begin_index,
                       end_index,
                       latitude_column,
                       latitude_value,
                       longitude_column,
                       longitude_value,
                       zoom):
    df = dataframe[begin_index:end_index]
    location = [latitude_value,longitude_value]
    plot = folium.Map(location=location,zoom_start=zoom)
    for i in range(0,len(df)):
        popup = folium.Popup(str(df.VILLAGE[i:i+1]))
        folium.Marker([df[latitude_column].iloc[i],df[longitude_column].iloc[i]],popup=popup).add_to(plot)
    return(plot)
```

```python
file_name = 'education-fac/Education_Facilities_3318.csv'
file_path = os.path.join(base_dir,file_name)
dataframe = pd.read_csv(file_path)
# plot all points instead of only 500 points for final draft
plot_points_on_map(dataframe,0,500,'LAT',-1.9403,'LONG',29.8739,9)
```
