The 2022 State of Open Data survey developed between Springer Nature, figshare and Digital Science. 
 - Attitudes and behaviours of research authors to the sharing, reuse and redistribution of openly available data.
 - For more detail, see https://www.springernature.com/gp/researchers/campaigns/state-of-open-data.

```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel('../input/state-of-open-data-2022/21295422/State of Open Data 2022_data to share.xlsx',sheet_name='Sheet1')
df.head()
```

```python
df['Q3.4'][0]
```

```python
value_counts = df['Q3.4'][1:].value_counts()[:-1]
value_counts
```

```python
fig = px.bar(value_counts[0:10], 
             x=value_counts[0:10].values, 
             y=value_counts[0:10].index,
             orientation='h',
             text=value_counts[0:10].values)
fig.update_layout(
            title=df['Q3.4'][0],
            xaxis_title='Count of responses',
            yaxis_title='Motivation')
fig.show()
```
