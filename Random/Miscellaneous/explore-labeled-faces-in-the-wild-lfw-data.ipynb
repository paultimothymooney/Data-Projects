```python
import numpy as np 
import matplotlib.pyplot as plt
import os
import PIL

height = 5; width = 5
folder_full_of_images = "../input/lfw-dataset/lfw-deepfunneled/lfw-deepfunneled/Abdullah_Gul/"
load_img = lambda filename: np.array(PIL.Image.open(f"../input/lfw-dataset/lfw-deepfunneled/lfw-deepfunneled/Abdullah_Gul/{filename}").resize((256, 256)))
_, axes_list = plt.subplots(height, width, figsize=(2*width, 2*height))
plt.subplots_adjust(wspace=0, hspace=0)
for axes in axes_list:
    for ax in axes:
        ax.axis('off')
        img = np.random.choice(os.listdir(folder_full_of_images))
        ax.imshow(load_img(img))
```
