**Map of Public WiFi Hotspots in NYC**

```python
import numpy as np 
import pandas as pd
import warnings
warnings.simplefilter("ignore")
import numpy as np
import pandas as pd
from scipy.special import boxcox
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
import plotly.plotly as py
import plotly.offline as offline
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import cufflinks as cf
from plotly.graph_objs import Scatter, Figure, Layout
from IPython.core import display as ICD
cf.set_config_file(offline=True)
train = pd.read_csv('../input/nyc-wi-fi-hotspot-locations.csv', nrows = 30_000)
train.head()
```

```python
target = train['BoroName']
data = [go.Histogram(x=target)]
layout = go.Layout(title = "Public Wifi Hotspots per Borough")
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
data = [go.Scattermapbox(
            lat= train['Latitude'] ,
            lon= train['Longitude'],
            customdata = train['BoroName'],
            mode='markers',
            text=train['Location'],
            marker=dict(
                size= 13,
                color = 'black',
                opacity = .8,
            ),
          )]
layout = go.Layout(autosize=False,
                   mapbox= dict(accesstoken="pk.eyJ1Ijoic2hhejEzIiwiYSI6ImNqYXA3NjhmeDR4d3Iyd2w5M2phM3E2djQifQ.yyxsAzT94VGYYEEOhxy87w",
                                bearing=0,
                                pitch=0,
                                zoom=13,
                                center= dict(
                                         lat=40.721319,
                                         lon=-73.887130),
                                #style= "mapbox://styles/shaz13/cjiog1iqa1vkd2soeu5eocy4i"
                               ),
                    width=1800,
                    height=1200, title = "Map of Public Wifi Hotspots in NYC")
fig = dict(data=data, layout=layout)
iplot(fig)
```
