**Map of Traffic Cameras in Chicago**

```python
import numpy as np 
import pandas as pd
import warnings
warnings.simplefilter("ignore")
import numpy as np
import pandas as pd
from scipy.special import boxcox
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
import plotly.plotly as py
import plotly.offline as offline
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import cufflinks as cf
from plotly.graph_objs import Scatter, Figure, Layout
cf.set_config_file(offline=True)
train = pd.read_csv('../input/speed-camera-locations.csv', nrows = 30_000)
train2 = pd.read_csv('../input/red-light-camera-locations.csv', nrows = 30_000)
train.head()
```

```python
print("Total Number of Speed Cameras: {}".format(train.shape[0]))
print("Total Number of Red Light Cameras: {}".format(train2.shape[0]))
```

```python
data = [go.Scattermapbox(
            lat= train['LATITUDE'] ,
            lon= train['LONGITUDE'],
            customdata = train['ADDRESS'],
            mode='markers',
            text=train['ADDRESS'],
            marker=dict(
                size= 20,
                color = 'black',
                opacity = .8,
            ),
          )]
layout = go.Layout(autosize=False,
                   mapbox= dict(accesstoken="pk.eyJ1Ijoic2hhejEzIiwiYSI6ImNqYXA3NjhmeDR4d3Iyd2w5M2phM3E2djQifQ.yyxsAzT94VGYYEEOhxy87w",
                                bearing=0,
                                pitch=0,
                                zoom=10,
                                center= dict(
                                         lat=41.881900,
                                         lon=-87.325808),
                                #style= "mapbox://styles/shaz13/cjiog1iqa1vkd2soeu5eocy4i"
                               ),
                    width=1800,
                    height=1200, title = "Speed Cameras in Chicago")
fig = dict(data=data, layout=layout)
iplot(fig)
```
