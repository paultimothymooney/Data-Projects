**Map of Bigfoot Sightings in America**

```python
import numpy as np 
import pandas as pd
import warnings
warnings.simplefilter("ignore")
from scipy.special import boxcox
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import plotly
import plotly.plotly as py
import plotly.offline as offline
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
import cufflinks as cf
from plotly.graph_objs import Scatter, Figure, Layout
cf.set_config_file(offline=True)
train = pd.read_csv('../input/bfro-report-locations.csv')
```

```python
print("Total Number of Bigfoot Sightings: {}".format(train.shape[0]))
data = [go.Scattermapbox(
            lat= train['latitude'] ,
            lon= train['longitude'],
            mode='markers',
            text=train['title'],
            marker=dict(
                size= 12,
                color = 'black',
                opacity = .8,
            ),
          )]
layout = go.Layout(autosize=False,
                   mapbox= dict(accesstoken="pk.eyJ1Ijoic2hhejEzIiwiYSI6ImNqYXA3NjhmeDR4d3Iyd2w5M2phM3E2djQifQ.yyxsAzT94VGYYEEOhxy87w",
                                bearing=0,
                                pitch=0,
                                zoom=3,
                                center= dict(
                                         lat=38.581900,
                                         lon=-79.325808),
                                #style= "mapbox://styles/shaz13/cjiog1iqa1vkd2soeu5eocy4i"
                               ),
                    width=1800,
                    height=1200, title = "Bigfoot Sightings in America")
fig = dict(data=data, layout=layout)
iplot(fig)
```
