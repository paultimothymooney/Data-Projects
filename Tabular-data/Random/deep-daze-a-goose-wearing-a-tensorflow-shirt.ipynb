

**Deep Daze is an open source tool for text to image generation.**  

[Deep Daze](https://github.com/lucidrains/deep-daze) is based off of [CLIP](https://github.com/openai/CLIP) and [Siren](https://arxiv.org/abs/2006.09661) and has some similarities to [Deep Dream](https://www.tensorflow.org/tutorials/generative/deepdream). For more detail, see https://github.com/lucidrains/deep-daze.

```python
from deep_daze import Imagine

phrase = 'a goose wearing a tensorflow shirt'

imagine = Imagine(
    text = phrase,
    image_width=512,
    num_layers = 16,
    save_every=1000,
    save_progress=True,
    save_date_time=True,   
    batch_size=16,
    gradient_accumulate_every=2
)
imagine()
```
